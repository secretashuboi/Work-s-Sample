 Local Outlier factors

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons


x, y = make_moons(n_samples=300, noise=0.1)
```

```python
x.shape
```

```python
# we are adding some outliers to the x

x_outliers =  np.random.uniform(low=-3, high=3, size=(20, 2))
x = np.vstack((x, x_outliers))
```

```python
x.shape
```

```python
out = LocalOutlierFactor(n_neighbors = 20 , contamination = 0.1)
y_pred = out.fit_predict(x)
```

```python
y_pred
```

```python
# Separate normal data points>>inliers and outliers

x_inliers = x[y_pred == 1]
x_outlier = x[y_pred == -1]

# Plotting
plt.title("Local Outlier Factor (LOF) on Moons Data")
plt.scatter(x_inliers[:, 0], x_inliers[:, 1], color='b', s=20, edgecolor='k', label='Normal data points')
plt.scatter(x_outliers[:, 0], x_outliers[:, 1], color='r', s=20, edgecolor='k', label='Outliers')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
adft = adfuller(stock_data["Close"])
```

```python
adft
#null hypotheis of adf test is time series is non stationary, pvalue<0.05, reject null hypothesis
#here pvalue> 0.05, it means series is non stationary
```

```python
#from the test it is clear that the data is non stationary

pd.Series(adft[0:4], index = ["test_statistics", "p-value", 'lag', 'no of data points'])
```

```python
# all the above things , we can also visualise them

def test_stationarity(timeseries):
    # Determing rolling statistics
    rolmean = timeseries.rolling(48).mean() # rolling mean
    rolstd = timeseries.rolling(48).std() # rolling standard deviation
    # Plot rolling statistics:
    plt.figure(figsize = (18,8))
    plt.grid('both')
    plt.plot(timeseries, color='blue',label='Original', linewidth = 3)
    plt.plot(rolmean, color='red', label='Rolling Mean',linewidth = 3)
    plt.plot(rolstd, color='black', label = 'Rolling Std',linewidth = 4)
    plt.legend(loc='best', fontsize = 20, shadow=True,facecolor='lightpink',edgecolor = 'k')
    plt.title('Rolling Mean and Standard Deviation', fontsize = 25)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.show(block=False)

    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    # hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
```

```python
test_stationarity(stock_data["Close"])
```
